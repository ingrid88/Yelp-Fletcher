### Take json and turn into panda 

```python
def json_to_panda(filename):
    """
    Input: directory/filename
        each row of file is in json form
    Output: pandas dataframe without deeply nested jsons used
    """
    with open(filename, 'rb') as f:
        data = f.readlines()
        data = map(lambda x: x.rstrip(), data)
        data_json_str = "[" + ','.join(data) + "]"
        data_df = pd.read_json(data_json_str)
        
    return data_df
```
